# **NLP**

```python
from sklearn.feature_extraction.text import CountVectorizer
# CountVectorizer used to convert text into matrix of tokens counts
# splits the data into individual words
# convert text into numerical value
```

```python
cv=CountVectorizer(max_features=10000,stop_words='english')
# max_feature -> used to specify max number of frequent words
# max_feature -> vectorizer will only consider 10,000 most frequent words in the corpus
# stop_words -> It is used to specify the words that should be ignored in the process of tokenization
# stop_words ->English words like is,the,or would be ignored
cv
```

```python
#EXAMPLE
# X=cv.fit_transform(documents)  ->It will convert the document into the words and then into numerical value
```

```python
matrix=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
#.values ->Convert the value in numpy array
# astype('U')->Converts the numpy array into Unicode string
# This step is often necessary to ensure that all values are treated as strings by downstream functions or libraries,
# such as CountVectorizer in scikit-learn.
# When working with data that contains different types, such as integers, floats,
#or object types, it's common to convert them to a uniform data type to ensure consistent processing.
# to.array() -> Convert the sparse matrix to dense Numpy array
# sparse matrix -> majority of elements is zero
# dense matrix -> majority of elements are non-zero
```

```python
matrix.shape
```

**USING COSINE SIMILARITY**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(matrix)
```

```python
similarity
```

```python
new_data[new_data['title']=="The Godfather"].index[0]
```

```python
# EXAMPLE
  # sorted(...,reverse=True,key= lambda .....)
  # enumerate -> make index  for the list,tuple or whatever you have
  # sorted(..,reverse=True,...) ->Descending order
  # lambda arguments:expression  -> f= lambda x,y:x+y
```

```python
matrix[1]
```

```python
distance=sorted(list(enumerate(similarity[2])),reverse=True,key=lambda matrix:matrix[1])
distance
```

```python
distance=sorted(list(enumerate(similarity[2])),reverse=True,key=lambda matrix:matrix[1])
for i in distance[1:6]:
  print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda matrix:matrix[1])
    for i in distance[1:6]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
